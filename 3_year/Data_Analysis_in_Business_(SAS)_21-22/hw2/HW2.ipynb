```python
categorical = ['product_rk', 'store_location_rk', 'PROMO1_FLAG', 'AUTORIZATION_FLAG']
numeric = ['month','day_year','PRICE_REGULAR', 'PRICE_AFTER_DISC' ]
#'day','month','day_year',

column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(handle_unknown="ignore"), categorical),
    ('scaling', StandardScaler(), numeric)
])

regr = BaggingRegressor(n_estimators=8, random_state=0)

pipeline = Pipeline(steps=[
    ('ohe_and_scaling', column_transformer),
    ('regression', regr)
])
```

```python
pipeline.fit(X_train, y_train).score(X_test, y_test)
```

Посмотрим простую метрику качества

```python
pred_train = pipeline.predict(X_train)
pred_test = pipeline.predict(X_test)

print('Acuracy_train: ', np.mean(pred_train - y_train))
print('Acuracy_test: ', np.mean(pred_test - y_test))
```

Ошибка достаточно маленькая

```python
mean_squared_error(pred_train,y_train), mean_absolute_error(pred_train,y_train), r2_score(pred_train,y_train)
```

```python
mean_squared_error(pred_test,y_test), mean_absolute_error(pred_test,y_test), r2_score(pred_test,y_test)
```
